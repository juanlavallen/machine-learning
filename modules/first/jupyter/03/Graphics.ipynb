Gráficos de Linea
===

Los gráficos de linea podemos realizarlois a partir de:

* Dos grupos de datos, uno para el eje **X** y otro para el eje **Y**
* Un grupo de datos para el eje **Y**

````python
import matplotlib.pyplot as plt


"""
En lugar de especificar los datos pertenecientes a las dos coordenadas, presentamos un solo grupo
Matplotlib se encarga de asignarle un valor X entero creciente
"""

y = [1, 4, 9, 16, 25, 36, 49, 74]

plt.plot(y)
plt.show()
````

Gráfico de Dispersión
===
Para generar datos aleatorios se suele partir de un valor llamada **semilla**.
Para especificar una semilla utilizamos el método de numpy **numpy.random.seed(value)**

A partir de esta semilla podemos crear valores de dispersion mediante los metodos de numpy:

```
numpy.random.rand(numero_de_arrays_generados, elementos_de_dispersion)
```

#### Un ejemplo podria quedar asi:

````python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(5)
data = np.random.rand(2, 10)

plt.scatter(data[0], data[1])
plt.show()
````

````python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs


data_mb, features = make_blobs(n_samples=20, centers=4, random_state=3)
plt.scatter(data_mb[:,0], data_mb[:,1], marker='o', c=features)
plt.show()
````

| Parámetro     | Descripción   |
| ------------- |:-------------:|
| seed          | Semilla|
| n_samples(int)| El numero total de puntos divididos en partes iguales entre los grupos. Por default 100 |
| centros       | El numero de centros a generar o las ubicaciones de los centros fijos. Por default 3 |
| cluster_std(float)| La desviacion estandar de los grupos. El valor por defecto es de 1.0 |
| shyffle(boolean) | Mezclar la muestras, su valor por defecto es True |
| random_state(int or None) | Semilla utilizada para generar numeros aleatorios, si el valor es None, usa el valor de np.random |

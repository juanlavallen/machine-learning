Matplotlib
===

* **legend(loc=())** le da una posición específica dentro del gráfico
* **plt.plot()** este método crea las curvas del gráfico

#### Los parametros que recibe el método **plt.plot()** son:
  1. Los puntos que corresponden a la coordenada X
  2. Los puntos que corresponden a la coordenada Y
  3. El color de la línea o tipo de representacion para los puntos
  4. Atributo **label** con la descripción de la curva. (Para que salga debemos user el metodo **legend()**)

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 2, 100)

plt.plot(x, x, 'red', label='Lineal')
plt.plot(x, x**2, label='Cuadratica')
plt.plot(x, x**3, label='Cubica')

plt.legend()
```

Límites en Ejes + Descripción y Titulo
===
* **plt.ylim(min, max)** para el eje Y 
* **plt.xlim(min, max)** para el eje X
* **xlabel()** agrega una descripcíon en el eje X
* **ylabel()** agrega una descripcíon en el eje Y
* **title()** agrega un titulo al gráfico
* **grid()** agrega una grilla al gráfico

```python
plt.ylim(0, 10)
plt.xlim(0, 5)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Estructura de Gráfico')

plt.grid()
```
